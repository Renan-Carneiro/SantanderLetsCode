```python
#Para o segundo exercício, você deve criar um programa que realize uma cópia do arquivo "alunos.csv".
#Essa cópia deve ser um arquivo chamado "alunos_copia.csv".

with open('alunos.csv','r') as arquivo, open('alunos_copia.csv','w',newline='') as arquivo_copia:
    leitor = csv.reader(arquivo)
    escritor = csv.writer(arquivo_copia)
    for x in leitor:
        escritor.writerow(x)
```

```python
#Neste exercício você deve criar um novo arquivo chamado "alunos_media.csv".
#Esse novo arquivo é uma cópia de "alunos.csv" porém com uma coluna a mais chamada "Média"
#que vai abrigar os valores das médias das provas de cada aluno da lista.

#abri um arquivo pra ler e outro pra escrever
with open('alunos.csv','r') as arquivo, open('alunos_media.csv','w',newline='',encoding='utf-8') as arquivo_copia:
    leitor = csv.reader(arquivo)
    escritor = csv.writer(arquivo_copia)
    header=next(leitor) #tirei a primeira linha que sao cabeçalhos
    header.append('Média')#adicionei a coluna média ao cabeçalho
    escritor.writerow(header)#escrevi o cabeçário
    for x in leitor:
        media=((float(x[3])+float(x[4])+float(x[5])+float(x[6]))/4)
        x.append(media)
        escritor.writerow(x)
        
#verificando
with open('alunos_media.csv','r') as arquivo:
    leitor = csv.reader(arquivo)
    for x in leitor:
        print(x)
```

```python
#Utilize a URL "https://swapi.dev/api/people/4/" para fazer a requisição dos
#dados de Darth Vader e extraia as informações "name" (nome), "height"
#(altura), "mass" (massa) e "birth_year" (ano de nascimento) e imprima cada
#dado em uma linha.

import requests as req
url= "https://swapi.dev/api/people/4/"

requisicao= req.get(url)
print(requisicao.status_code) #verifica se deu certo
```

```python
dadosVader=requisicao.json() #guardando num dicionário
```

```python
nome= dadosVader['name']
altura= dadosVader['height']
massa= dadosVader['mass']
anonascimento=dadosVader['birth_year']

print(f'Seu nome é {nome}.')
print(f'Sua altura é de {float(altura)/100}m.')
print(f'Sua massa é de {massa} kg.')
print(f'Ele nasceu em {anonascimento}.')
```

```python
"""
5 - Em 2019 surgiram os primeiros casos de COVID-19 que se alastrou pelo mundo resultando numa pandemia.
A proposta deste exercício é utilizaruma API com informações de COVID-19 do mundo todo desde o início da
proliferação e descobrir que dia o Brasil confirmou o primeiro caso de Coronavírus.
Para fazer isso, utilize a URL "https://api.covid19api.com/country/brazil" em seu código. Ela retorna uma
lista de dicionários, onde cada dicionário traz informações através das chaves:
"ID", "Country", "CountryCode", "Province", "City","CityCode","Lat","Lon","Confirmed","Deaths", "Recovered", "Active" e "Date".
Utilize essas informações e retorne a data em que o Brasil confirmou o
primeiro caso de COVID.
"""
url ='https://api.covid19api.com/country/brazil'

reqCov= req.get(url)
print(reqCov.status_code) #verifica se deu certo
```

```python
dadosCovid=reqCov.json()
```

```python
for i in range(len(dadosCovid)): 
    if (dadosCovid[i]['Confirmed'] > 0):
        dataCaso=dadosCovid[i]['Date'][:10]
        print(f'O primeiro caso de Covid aconteceu na data: {dataCaso}')
        break #talvez fosse mais elegante se feito com um while pra não precisar dar break
```
